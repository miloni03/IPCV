```python
import cv2
import numpy as np
image1 = cv2.imread('b4.png')
image2 = cv2.imread('b5.png')
```

```python
weightedSumadd = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
cv2.imshow('Weighted Image', weightedSumadd)

```

```python
plt.imshow(image1)
```

```python
plt.imshow(image2)
```

```python
plt.imshow(weightedSumadd)
```

```python
sub = cv2.subtract(image1, image2)
```

```python
plt.imshow(sub)
```

```python
dest_and = cv2.bitwise_and(image2, image1, mask = None)
```

```python
cv2.imshow('Bitwise And', dest_and)
```

```python
plt.imshow(dest_and)
```

```python
dest_or = cv2.bitwise_or(image2, image1, mask = None)
```

```python
plt.imshow(dest_or)
```

```python
dest_xor = cv2.bitwise_xor(image1, image2, mask = None)
```

```python
plt.imshow(dest_xor)
```

```python
dest_not1 = cv2.bitwise_not(image1, mask = None)
dest_not2 = cv2.bitwise_not(image2, mask = None)
```

```python
plt.imshow(dest_not1)
```

```python
plt.imshow(dest_not2)
```
