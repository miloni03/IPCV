```python
# Contrast Stretching
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img5.ppm  ", 0)
```

```python
plt.imshow(img)
```

```python
def contrast_stretch(r,a,b,l,m,n):
    v=l*a
    w=m*(b-a)+v
    if(r>0 and r<=a):
        return(l*r)
    elif(r>a and r<=b):
        return(m*(r-a)+v)
    elif(r>b and r<(L-1)):
        return(n*(r-b)+w)
    
    
    
```

```python
a=100
b=150
l=0.6
m=1.5
n=0.6
L=256
```

```python
plt.imshow(img,cmap="gray")
```

```python
print("Min:",img.min())
```

```python
print("Max:",img.max())
```

```python
import numpy as np
ob=np.vectorize(contrast_stretch)
```

```python
ob1=ob(img,a,b,l,m,n)
```

```python
plt.imshow(ob1,cmap="gray")
```

```python
print(ob1.min())
```

```python
print(ob1.max())
```

```python
# Taking Reverse
a=100
b=150
l=1.6
m=0.5
n=1.6
L=256
```

```python
ob2=ob(img,a,b,l,m,n)
```

```python
plt.imshow(ob2,cmap="gray")
```

```python
ob2.min()
```

```python
ob2.max()
```

```python
# Log transformation
```

```python
c=200
op=c*np.log(img+1)
op=np.array(op)
plt.imshow(op,cmap="gray")
```

```python
# Power Transformation
```

```python
g1=0.5
g2=2.5
powert=img**g1
plt.imshow(powert,cmap="gray")
```

```python
powert=img**g2
plt.imshow(powert,cmap="gray")
```

Conclusion: Hence, by doing contrast stretching, dynamic range of the image increases and thus affecting its contrast.
During log transformation, the dark pixels in an image are expanded as compare to the higher pixel values. The higher pixel values are kind of compressed in log transformation.
In Power law, greater the gamma value more darker is the output image.
