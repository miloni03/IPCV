```python
# Importing Modules
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import numpy as np
import math
import random
```

```python
#Reading image
img = cv2.imread('/content/img3.png')
```

```python
# Displaying Image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Displaying Grey Scale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img , cmap = "gray")
```

```python
# Displaying Shape
x , y = (img.shape)
print(x)
print(y)
```

```python
# High pass filter
```

```python
# Mask
filter = np.ones([3, 3], dtype = int)
filter = - filter
filter[1,1] = - 8 * filter[1,1]
print(filter)
```

```python
# High Pass Filter
hp_filter = np.zeros([x, y])

for i in range(1, x-1):
	for j in range(1, y-1):
		temp = (img[i-1, j-1]*filter[0, 0] + img[i-1, j]*filter[0, 1] + 
          img[i-1, j + 1]*filter[0, 2] + img[i, j-1]*filter[1, 0] + 
          img[i, j]*filter[1, 1] + img[i, j + 1]*filter[1, 2] + 
          img[i + 1, j-1]*filter[2, 0] + img[i + 1, j]*filter[2, 1] +
          img[i + 1, j + 1]*filter[2, 2])
		
		hp_filter[i, j]= temp
```

```python
fig=plt.figure(dpi = 150)
fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(hp_filter,cmap='gray')
plt.axis("off")
plt.title("High Pass Filtered")
```

```python
# High Boost Filter
```

```python
a = 1.5
```

```python
filter = np.ones([3, 3], dtype = int)
filter = - filter
filter[1,1] = -( 8 + a )* filter[1,1]
```

```python
hb_filter = np.zeros([x, y])

for i in range(1, x-1):
	for j in range(1, y-1):
		temp = (img[i-1, j-1]*filter[0, 0] + img[i-1, j]*filter[0, 1] + 
          img[i-1, j + 1]*filter[0, 2] + img[i, j-1]*filter[1, 0] + 
          img[i, j]*filter[1, 1] + img[i, j + 1]*filter[1, 2] + 
          img[i + 1, j-1]*filter[2, 0] + img[i + 1, j]*filter[2, 1] +
          img[i + 1, j + 1]*filter[2, 2])
		
		hb_filter[i, j]= temp
```

```python
fig=plt.figure(dpi = 150)
fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(hb_filter,cmap='gray')
plt.axis("off")
plt.title("High Boost Filtere")
```

**Conclusion:**

**i. In  High Pass Filtering, it highlights the fine details in an image. It is use to enhance the image and removes the background.**

**ii. In High Boost Filtering, it enhances the image and also retains the background.**
